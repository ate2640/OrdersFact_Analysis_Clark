### Timespan for order shipments
There are two columns, namely Order Date and Ship Date, whose difference tells us the timespan, in days, between the Order
Placement and Order Delivery. I will create an additional column to see what the timespans are per item:

```python
OrdersFact.head()
```

```python
Order_TimeSpan = OrdersFact["Ship Date"] - OrdersFact["Order Date"]
OrdersFact["Order_Timespan"] = Order_TimeSpan
```

```python
OrdersFact.head()
```

Now that we've successfully included a timespan column to see the number of days between Order Date and Ship Date, let's 
view important information pertaining to the delivery time:

```python
print(OrdersFact.Order_Timespan.describe())
print("The fastest delivery is same day delivery, and the longest delivery took 364. The average delivery takes 107 days to complete.")
```

```python
same_day_delivery = OrdersFact[OrdersFact['Order_Timespan']==0]
same_day_delivery
```
