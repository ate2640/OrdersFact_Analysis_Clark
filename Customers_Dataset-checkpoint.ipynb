### As this is a customer-based dataset, let's see the breakdown of Account Status:

```python
customers_dataset['Account Status'].unique()
```

```python
customers_dataset.groupby('Account Status')['Customer ID'].count().plot.barh()
print(customers_dataset.groupby('Account Status')['Customer ID'].count())
print(round(12847*100/18372,2), "% of customers are active")
print(round(920*100/18372,2), "% of customers canceled")
print(round(1834*100/18372,2), "% of customers are inactive")
print(round(25*100/18372,2), "% of customers are new")

print(round(2746*100/18372,2), "% of customers are Past due")
```

### In addition to Account Status, let's view the breakdown of Customer Tier:

```python
customers_dataset.groupby('Customer Tier')['Customer ID'].count().plot.barh()
print(customers_dataset.groupby('Customer Tier')['Customer ID'].count())
print(round(5511*100/18372,2), "% of customers are Bronze")
print(round(4592*100/18372,2), "% of customers are Gold")
print(round(1837*100/18372,2), "% of customers are Platinum")
print(round(1835*100/18372,2), "% of customers are Platinum Elite")
print(round(4597*100/18372,2), "% of customers are Silver")
```

### Customers, by gender:

```python
customers_dataset.groupby('Gender')['Customer ID'].count().plot.barh()
print(customers_dataset.groupby('Gender')['Customer ID'].count())
```

```python
customers_dataset['Age'].unique().size
```

### Another essential finding of customer data is age demographics:

```python
customers_dataset.groupby('Age')['Customer ID'].count().plot.barh()
#df.groupby('Payment Mode').Country.count().plot.barh()
print(customers_dataset.groupby('Age')['Customer ID'].count())
print(customers_dataset.Age.describe())
```

```python
print("Customers come from", customers_dataset.City.unique().size, "cities and",
customers_dataset.Country.unique().size, "countries")
```

### After these findings, I then subset the data, by country, beginning with the USA:

```python
print(customers_dataset.groupby('Country')['Customer ID'].count())
customers_dataset.groupby('Country')['Customer ID'].count().plot.barh()
print(round(3567*100/18372,2), "% of customers are from Australia")
print(round(1638*100/18372,2), "% of customers are from Canada")
print(round(1770*100/18372,2), "% of customers are from France")
print(round(1734*100/18372,2), "% of customers are from Germany")
print(round(1908*100/18372,2), "% of customers are from the UK")
print(round(7755*100/18372,2), "% of customers are from the USA")
```

```python
customers_USA = customers_dataset[customers_dataset["Country"] == "United States"]
```

#### Density map of the USA and the location of its clients

```python
customers_USA.plot(kind="scatter", x="Longitude", y="Latitude", alpha=0.1)
```

```python
customers_USA
```

### The USA clients come from 35 states 

```python
print("Number of customers per state in the USA:")
print(customers_USA.groupby('State/Province')['Customer ID'].count())
print("California, Oregon and Washington account for", round((4261+1015+2179)*100/7755,2), "% of existing American clients")
print("and", round((4261+1015+2179)*100/18371,2), "% of all clients in the world" )
```

### After analyzing the USA's dataset, I then subset the UK's portion for further analysis:

```python
customers_UK = customers_dataset[customers_dataset["Country"] == "United Kingdom"]
customers_UK
```

### England is the only State/Province in the UK from this dataset, so I analyzed the dataset by city:

```python
print("Number of customers per city in the UK:")
print(customers_UK.groupby('City')['Customer ID'].count())
print("London and York are share the largest proportion of clientele, accounting for", round((420+115)*100/1908,2),"% of the UK's customer base")
```

### Analysis of Canada's customer base:

```python
print("Number of customers per province in Canada:")
print(customers_canada.groupby('State/Province')['Customer ID'].count())
print("With", round((1537)*100/1638,2),"% of the Canada's customer base British Columbia by far has the largest share of customers")
```
